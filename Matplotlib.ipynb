```python
#x^3
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="red",linewidth=5,label="x^3")
plt.plot(x,y2,color="black",linewidth=5,label="x^2")
# plt.axis([0,10,0,150])
plt.grid()
plt.title("Demo")

plt.ylabel("y")
plt.xlabel("x")

plt.legend()

plt.text(3,60,"Check",fontsize=14,color="blue")

plt.show()
```

## bubble-graph

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
```

```python
plt.scatter(year,salary,s=population,c='r')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c='r',alpha=0.6)
plt.title("Bubble Chart")
plt.show()
```

```python
#array of clors so that every bubble gets different color
c=np.arange(len(year))

plt.scatter(year,salary,s=population,c=c,alpha=0.6,edgecolor='black')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.6,edgecolor='black',marker='^')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.6,edgecolor='black',marker='^')
plt.title("Bubble Chart")
plt.text(year[0]+0.1,salary[0]+0.1,'Population= {pop}'.format(pop=population[0]))
plt.show()
```

## pie-graphs

```python
sizes=[3,4,6,2]
labels=["A","B","C","D"]
colors=["blue","cyan","red","pink"]
plt.title("My pie graph")
explode=[0.1,0,0,0]
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.2f%%",counterclock=False,startangle=80)
plt.axis("equal")
plt.show()
```

```python
# size of bin=(maxno-minno)/no of bins
```
